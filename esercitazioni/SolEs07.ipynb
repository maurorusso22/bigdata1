Basandovi sugli esempi visti a lezione provate qualche query sulle collections di sample_mflix.

```python
# cerca i film rated R con Meryl Streep oppure Charlize Theron
query = {"$or" : [{"cast":"Meryl Streep"}, {"cast" : "Charlize Theron"}], "rated":"R"}
prop = {"_id":0, "title":1, "year":1, "cast":1, "rated":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

```python
# cerca i film con Charlize Theron con anno <= 2015
query = {"cast" : "Charlize Theron", "year":{"$lte":2015}}
prop = {"_id":0, "title":1, "year":1, "cast":1, "rated":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

```python
# cerca i film che non sono di type movie e hanno num_mflix_comments > 0
query = {"type": {"$ne" : "movie"}, "num_mflix_comments": {"$gt" : 0}}
prop = {"_id":0, "title":1, "type":1, "cast":1, "num_mflix_comments":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

```python
# cerca i film che hanno un imdb rating > 9
query = {"imdb.rating": {"$gt" : 9}}
prop = {"_id":0, "title":1, "type":1, "cast":1, "imdb":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

Esempi con aggregazione

```python
# conta numero totale di film
group_stg = {"$group":{"_id":"null", "count":{"$sum":1}}}
res = db.movies.aggregate([group_stg])
list(res)
```

```python
# conta numero di film con Meryl Streep
match_stg = {"$match" : {"cast":"Meryl Streep"}}
group_stg = {"$group":{"_id":"null", "count":{"$sum":1}}}
res = db.movies.aggregate([match_stg, group_stg])
list(res)
```

```python
# conta numero di film americani per ciascun rating
match_stg = {"$match" : {"countries":"USA"}}
group_stg = {"$group":{"_id":"$rated", "count":{"$sum":1}}}
res = db.movies.aggregate([match_stg, group_stg])
list(res)
```

```python
# trova i 3 film italiani che hanno il cast più numeroso
# nota: nel filtro iniziale elimino i film per cui non c'è il cast
match_stg = {"$match" : {"countries":"Italy", "cast":{"$exists":"true"}}}
project_stg = {"$project" : {"_id":0, "title":1, "countries":1, "ncast": {"$size":"$cast"}}}
sort_stg = {"$sort": {"ncast":-1}}
limit_stg = {"$limit":5}
res = db.movies.aggregate([match_stg, project_stg, sort_stg, limit_stg])
list(res)
```
