##Caricate il musk dataset ed eliminate le prime colonne non numeriche


```python
import pandas as pd

labels=['molecule', 'conformation']+list(range(166)) +['class']
df = pd.read_csv('http://www.lacascia.it/bd2/clean2.data', names=labels)
```

```python
df
```

##Effettuate una classificazione k-NN su tutto il dataset con split fra train e test 75/25


```python
data = df.iloc[:, 2:-1].to_numpy(copy=True)
target = df.iloc[:, -1].to_numpy(copy=True)
print(target.shape, data.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=11)
knn = KNeighborsClassifier()
knn.fit(X=X_train, y=y_train)

print(f'{knn.score(X_test, y_test):.2%}')
```

##Fate riduzione di dimensionalità con PCA e classificate nuovamente con k-NN


```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.9)
data_n = pca.fit(data).transform(data)
print(pca.n_components_)

Xn_train, Xn_test, yn_train, yn_test = train_test_split(data_n, target, random_state=11)
knn2 = KNeighborsClassifier()
knn2.fit(X=Xn_train, y=yn_train)

print(f'{knn2.score(Xn_test, yn_test):.2%}')
```

##Sul dataset a dimensionalità ridotta fate un compionamento senza replacement del 40% ed effettuate la classificazione con k-NN sul dataset campionato (split sempre 75/25)


```python
import numpy as np
n = data_n.shape[0]
dataset = np.append(data_n, target.reshape(n,1), axis=1)
print(dataset.shape)

dataset_c = dataset[np.random.choice(n, int(n*0.4), replace=False), :]
print(dataset_c.shape)
data_c = dataset_c[:, 2:-1]
target_c = dataset_c[:, -1]

Xc_train, Xc_test, yc_train, yc_test = train_test_split(data_c, target_c, random_state=11)
knn3 = KNeighborsClassifier()
knn3.fit(X=Xc_train, y=yc_train)

print(f'{knn3.score(Xc_test, yc_test):.2%}')
```

#Esercizio 2 - email spam detector
## Importate in un DataFrame il dataset che contiene 9997 email etichettate (1=spam, 0=non_spam)

```python
import pandas as pd

labels=['class', 'text']
df = pd.read_csv('http://www.lacascia.it/bd2/combined_data_small.csv', names=labels)
```

```python
print(df.shape)
df.iloc[:,0].value_counts()
```

##Determinate il dizionario delle parole che compaiono almeno 500 volte in tutto il dataset (nota: dovrebbero essere 650)

```python
dict = {}
for row in df.iloc[:, 1]:
  words = row.split()
  for w in words:
    if w in dict:
      dict[w]+=1
    else:
      dict[w]=1
print(f'Parole totali: {len(dict.keys())}')

freq_words = {key: val for key,val in dict.items() if dict[key]>500}
print(f'Parole che compaiono almeno 500 volte: {len(freq_words)}')
```

## Costruite una matrice NumPy 9997x650 che contiene per ciascuna riga la frequenza delle 650 parole più comuni nella  mail (ogni riga corrisponde a una mail del dataset)

```python
import numpy as np
data = np.zeros((9997, 650))
i=0
for row in df.iloc[:, 1]:
  words = row.split()
  for k in freq_words.keys():
    freq_words[k]=0
  for w in words:
    if w in freq_words:
      freq_words[w]+=1
  for j, val in enumerate(freq_words.values()):
    data[i, j] = val
  if i%500 == 0:
    print(i)
  i+=1
print(data.shape)
```

##Utilizzando la prima colonna del DataFrame come *target* effettuate una classificazione con k-NN e valutatene l'accuratezza

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

data_s = data[:, 1:]
target_s = data[:, 0]

Xs_train, Xs_test, ys_train, ys_test = train_test_split(data_s, target_s, random_state=11)
knns = KNeighborsClassifier()
knns.fit(X=Xs_train, y=ys_train)
print(f'{knns.score(Xs_test, ys_test):.2%}')
```

##Effettuate una riduzione di dimensionalità con PCA ed effettuate nuovamente la classificazione. Commentate l'accuratezza ottenuta.

```python
from sklearn.decomposition import PCA
pca_s = PCA(n_components=0.85)
data_sn = pca_s.fit(data_s).transform(data_s)
print(pca_s.n_components_)

Xsn_train, Xsn_test, ysn_train, ysn_test = train_test_split(data_sn, target_s, random_state=11)
knns = KNeighborsClassifier()
knns.fit(X=Xsn_train, y=ysn_train)

print(f'{knns.score(Xsn_test, ysn_test):.2%}')
```
