# Insert

```
article = {"_id": "n1",
           "author": "Derrick Mwiti",
           "about": "Introduction to MongoDB and Python",
           "tags": ["mongodb", "python", "pymongo"]}
```

```
print(f'First article key is: {result.inserted_id}')
```

# Update

```
query = { "author": "Derrick Mwiti" }
```

```
new_author = { "$set": { "author": "John David" } }
```

```
db.articles.update_many(query, new_author)
```

```
for doc in db.articles.find({},{ "_id": 0, "author": 1, "about": 1}):
  print(doc)
```
