```python
from scipy.spatial.distance import minkowski
from scipy.spatial.distance import chebyshev
from matplotlib.pyplot import plot
import numpy as np
```

```python
x = (1, 2.6, 4, 4.5, 5, 4.5, 4, 3.2, 1.8, 1.2, 1)
y = (1, 4.5, 5, 4, 2, 1)
#y = (1, 4, 4.45, 4, 2, 1)
```

```python
plot(x)
plot(y)
```

```python
DTW = np.zeros((len(x)+1, len(y)+1))
DTW[0, 1:] = np.ones(len(y))*1000
DTW[1:, 0] = np.ones(len(x))*1000
DTW
```

```python
for i in range(1, len(x)+1):
    for j in range(1, len(y)+1):
        DTW[i, j] = minkowski([x[i-1]], [y[j-1]], 2) + min(DTW[i, j-1], DTW[i-1, j], DTW[i-1, j-1])
DTW[len(x), len(y)]
```
