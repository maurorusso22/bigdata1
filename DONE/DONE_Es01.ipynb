Calcolare media varianza e deviazione standard e verificare che il risultato sia coerente con quello del modulo statistics

```python
from statistics import mean
from statistics import pvariance
from statistics import pstdev
from math import pow, sqrt
import numpy as np

array = [34, 98, 23, 57, 64, 87, 72, 33, 48, 29]

def my_mean(a):
  sum = 0
  for x in a:
    sum += x
  return sum / len(a)

def my_variance(a):
  m = mean(a)
  sum = 0
  for x in a:
    sum += (x - m) * (x - m)
  return sum / len(a)
  # return sum((xi - m) ** 2 for xi in a) / len(a)

print(f'mean: statistics: {mean(array)} --- my_mean: {my_mean(array)}')
print(f'variance: statistics: {pvariance(array)} --- my_variance: {my_variance(array)}')
print(f'standard dev: statistics: {pstdev(array)} --- my_variance: {sqrt(my_variance(array))}')
print(f' numpy values: {np.mean(array)}, {np.var(array)}, {np.std(array)}')
```

4.9 (Degrees to Radians) Implement a radians function that returns the radian equiv- alent of a degree. Use the following formula:
        rad = degrees * pi/180
Use this function to print a chart showing the radian equivalent of all degrees ranging from 1° to 180°. Use two digits of precision for the results. Print the outputs in a neat tabular format.

```python
import math
from prettytable import PrettyTable

def dtr(degree):
  return degree * (math.pi)

t = PrettyTable(['Degrees', 'Radians'])
d = 0
while (d <= 180):
  t.add_row([d, dtr(d)])
  d += 15

print(t)
```

```python
# other option
from tabulate import tabulate
import math

def dtr(degree):
  return degree * (math.pi)

array = []
d = 0
while (d <= 180):
  array.append([d, dtr(d)])
  d += 15

print(tabulate(array, headers=['Degrees', 'Radians']))
```

5.7 (Duplicate Elimination) In organizations, a list of email addresses is often compiled for marketing purposes. However, duplicate email addresses need to be removed from this list. Write a function that receives a list and returns a list containing only unique values. Test your function with a list of email addresses.

```python
emails = ['a@g.c', 'b@g.c', 'c@g.c', 'b@g.c', 'd@g.c', 'a@k.t', 'a@g.c']
def remove_duplicates(array):
  result = []
  for x in array:
    if (x not in result):
      result.append(x)
  return result

f = remove_duplicates(emails)
print(f)
```

5.16 (Sorting User IDs) For each new user, a system generates a user ID consisting of a randomly chosen letter and two random numbers. Write a script that randomly generates a list containing 8000 user IDs using random letters from the range "a" through "f" and numbers from the range 1 to 4. Subsequently, remove all duplicate entries from the list and sort this list in descending order.

```python
from random import randint

def genID():
  l = ['a', 'b', 'c', 'd', 'e', 'f']
  id = l[randint(0, len(l) - 1)] + str(randint(0,9)) + str(randint(0,9))
  return id

def remove_duplicates(array):
  result = []
  for x in array:
    if (x not in result):
      result.append(x)
  return result

ids = [genID() for x in range(8000)]
fids = remove_duplicates(ids)
print(len(ids), len(fids))
fids.sort(reverse=True)
print(fids[:10])

```

5.19 (Finding the People with a Specified Last Name) Create a list of tuples containing first and last names. Use filter to locate the tuples containing the last name Jones. En- sure that several tuples in your list have that last name.

```python
from random import randint

def genNameSurname():
  n = ['Marc', 'Paul', 'John', 'Brad', 'Anthony', 'Morgan', 'Xavier']
  s = ['Jones', 'Depp', 'Pitt', 'Smith', 'Roses', 'Logan', 'James']
  ns = (n[randint(0, len(n) - 1)], s[randint(0, len(s) - 1)])
  return ns

people = [genNameSurname() for x in range(100)]
jones = list(filter(lambda x: x[1] == 'Jones', people))
print(len(people), len(jones))
print(jones[:5])
```
